# dataset used: train.csv

```python
import pandas as pd 
df = pd.read_csv('train.csv')
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
X = df.drop(columns=['Id', 'SalePrice'])
y = df['SalePrice']
```

```python
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object']).columns
```

```python
for col in num_cols:
    X[col] = X[col].fillna(X[col].median())
for col in cat_cols:
    X[col] = X[col].fillna('None')
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
import numpy as np 
y_train_log = np.log1p(y_train)
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1.0)
model.fit(X_train, y_train_log)
```

```python
log_predictions = model.predict(X_val)
final_predictions = np.expm1(log_predictions)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, root_mean_squared_error
val_mae = mean_absolute_error(y_val, final_predictions)
val_r2 = r2_score(y_val, final_predictions)*100
val_rmse_log = root_mean_squared_error(np.log1p(y_val), log_predictions)
```

```python
print("--- Model Performance ---")
print(f"Mean Absolute Error (MAE): ${val_mae:,.2f}")
print(f"R² Score: {val_r2:.4f}%")
print(f"Log-Scale RMSE: {val_rmse_log:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_val, y=final_predictions, alpha=0.6, color='purple', edgecolor='w')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--', lw=2, label='Perfect Prediction')

plt.title('Actual vs. Predicted House Prices (Log-Transformed Ridge)', fontsize=14)
plt.xlabel('Actual SalePrice ($)', fontsize=12)
plt.ylabel('Predicted SalePrice ($)', fontsize=12)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f"${x:,.0f}"))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f"${x:,.0f}"))
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

### Predicting House Price 

```python
new_house = X_val.iloc[[0]] 
predicted_log_price = model.predict(new_house)

predicted_usd_price = np.expm1(predicted_log_price)
print(f"Predicted Value: ${predicted_usd_price[0]:,.2f}")
```
